```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import (accuracy_score, classification_report, confusion_matrix, ConfusionMatrixDisplay, precision_score, recall_score, f1_score)

print("All libraries Import successfully!")
```

```python
iris_df = pd.read_csv("Iris.csv")
```

```python
iris_df.head()
```

```python
# Checking for missing values
# iris_df.info() # or
print("Missing Values")
print(iris_df.isnull().sum())
print("Data types: ", iris_df)
```

```python
# Checking target Class balance
# Imp: imbalanced classes can mislead accuracy
print("Target Distribution")
print(iris_df["Species"].value_counts())
print()

# Visualization
counts = iris_df["Species"].value_counts()
plt.figure(figsize = (7,8))
plt.bar(["Iris-setosa", "Iris-versicolor", "Iris-virginica"], counts.values, color = ["steelblue", "green", "orange"], edgecolor = "black")
plt.title("Iris flower species distribution")
plt.ylabel("Count")
for i, v in enumerate(counts.values):
    plt.text(i, v+1,str(v), ha="center", fontweight="bold")
plt.tight_layout()
plt.show()
```

```python
# Seperating Features and classifier(target)
X = iris_df.drop(columns = ["Id", "Species"])
y = iris_df["Species"]
```

```python
X.head()
```

```python
# Spliting data into train test
X_train, _, y_train, _ = train_test_split(
    X, y, test_size = 0.5, random_state = 42, stratify = y
)
X_test = X
y_test = y
```

```python
# Scaling using Standard Scaler
scaler = StandardScaler()

X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

```python
# Baseline model for logistic reression
model_baseline = LogisticRegression(C = 1.0, solver = "lbfgs", max_iter = 1000, random_state = 42)
model_baseline.fit(X_train_scaled, y_train)
y_pred = model_baseline.predict(X_test_scaled)
```

```python
#  Checking the accuracy score to check overfitt, undefitt
train_acc = model_baseline.score(X_train_scaled, y_train)
test_acc = model_baseline.score(X_test_scaled, y_test)

print(f"Train Accuracy : {train_acc}")
print(f"Test Accuracy : {test_acc}")
```

```python
# Evaluation
print("Recall score: ", recall_score(y_test, y_pred, average = 'macro'))
print("Accuracy score: ", accuracy_score(y_test, y_pred))
print("Precision score: ", precision_score(y_test, y_pred, average = 'macro'))
```

```python
# Confusion Matrix
cm = confusion_matrix(y_test,y_pred)
disp = ConfusionMatrixDisplay(confusion_matrix = cm, display_labels = model_baseline.classes_)
disp.plot(cmap="Blues")
plt.title("Baseline model Confusion Matrix")
plt.tight_layout()
plt.show()

# Reports
print("\nDetailed reports")
report = classification_report(y_test, y_pred)
print(report)
```
